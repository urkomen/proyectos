### Descripción

Voy a realizar un EDA sobre exoplanetas, su habitabilidad y plantear la paradoja de Fermi.
Creo que es un tema que fascina al público general porque conecta directamente con la famosa pregunta:  

“¿Hay vida ahí fuera?”

La clave de este análisis es no hablar de alienígenas ni marcianos, me centraré en los factores 
que afectan a la probabilidad de civilizaciones usando datos astronómicos reales.

Los **exoplanetas** me dan el material empírico: tamaños, temperaturas, distancias, tipos de estrella, zonas habitables. Con la **habitabilidad** definiré criterios cuantitativos: temperatura compatible con agua líquida, tipo de estrella estable, tamaño adecuado, etc. La **Paradoja de Fermi** me proporciona un marco conceptual: si hay tantos planetas potencialmente habitables, ¿por qué no vemos señales de vida inteligente?

Trataré de plantear el proyecto combinando datos, ciencia y filosofía de forma accesible y atractiva.

Hipótesis que tengo en la cabeza:
1. Los métodos actuales detectan preferentemente planetas grandes y calientes, no los más habitables.
2. Las estrellas más estables (actividad baja) tienen más planetas potencialmente habitables.
3. La mayoría de las estrellas cercanas no cumplen simultáneamente condiciones para la vida inteligente (edad, estabilidad, metalicidad, zona habitable,...)
4. La distancia media entre planetas potencialmente habitables es tan grande que la comunicación entre sí o viajes interestelares es muy improbable

Tal vez se me ocurra algo por el camino, pero esta es la idea principal que tengo en estos momentos
