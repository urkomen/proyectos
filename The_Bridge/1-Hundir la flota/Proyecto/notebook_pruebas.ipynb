```python
import random
import Barcos as bar


def generar_barco_aleatorio(tipo:int, tablero:tuple[int, int]):
    '''
    Genera aleatoriamente coordenadas de 
    un barco del tipo indicado
    '''
    
    while True:
        coords = []
        
        orientacion = random.choice(['H', 'V'])
        # Para que no se salga del tablero le ponemos l├нmites
        if orientacion == 'H':
            fil = random.randint(0, 9)
            col = random.randint(0, 10 - tipo)
         
            # Guardamos las coordenadas del barco
            for c in range(col, col + tipo):
                coords.append((fil, c))

            
        else: # orientacion == 'V'
            fil = random.randint(0, 10 - tipo)
            col = random.randint(0, 9)
            
            # Guardamos las coordenadas del barco
            for f in range(fil, fil + tipo):
                coords.append((f, col))

        # Validar que no se solapa con otros barcos
        for c in coords:
            if not tb.es_agua(tablero, c[0], c[1]):
                continue

        return coords
```

```python
def coords2index(coord: str):
    """
    Convierte una coordenada tipo 'A5' en ├нndices [fila, columna].
    A -> 0, B -> 1, ..., J -> 9
    1 -> 0, 2 -> 1, ..., 10 -> 9
    """
    coord = coord.strip().upper()
    fil = coord[0] # 'A'
    col = coord[1:] # '5' o '10'
    print(fil, col)
    for ind, valor in enumerate(tb.coordenadas):
        if valor == fil:
            fila_idx = ind
    col_idx = int(col) - 1

    return (fila_idx, col_idx)
```
